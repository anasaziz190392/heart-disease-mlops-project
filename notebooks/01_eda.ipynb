This notebook loads the raw Cleveland Heart Disease dataset, cleans it, and performs exploratory data analysis with professional visualizations.

```python
import sys, os
sys.path.append(os.path.join('..', 'src'))
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from data_prep import load_raw, clean, RAW_PATH
sns.set_theme(style='whitegrid')
%matplotlib inline
```

## 1. Load Raw Data

```python
raw_df = load_raw(RAW_PATH)
print(raw_df.shape)
raw_df.head()
```

## 2. Missing Values

```python
raw_df.isna().sum()
```

`ca` and `thal` have a handful of missing values (encoded as `?` in the raw file). We impute them with the column mode since they are ordinal/categorical codes.

## 3. Clean the Data

```python
df = clean(raw_df)
print(df.shape)
df.head()
```

## 4. Class Balance

```python
ax = df['target'].value_counts().sort_index().plot(kind='bar', color=['#4C72B0','#C44E52'])
ax.set_xticklabels(['No Disease','Disease'], rotation=0)
ax.set_title('Class Balance')
plt.show()
```

## 5. Numeric Feature Distributions

```python
numeric_cols = ['age','trestbps','chol','thalach','oldpeak']
df[numeric_cols].hist(figsize=(14,8), bins=20)
plt.suptitle('Numeric Feature Distributions')
plt.tight_layout()
plt.show()
```

## 6. Correlation Heatmap

```python
plt.figure(figsize=(11,9))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

## 7. Key Observations
- `cp` (chest pain type), `thalach` (max heart rate), `exang` (exercise induced angina), `oldpeak`, `ca`, and `thal` show the strongest association with the target.
- The dataset is moderately balanced (roughly 54%/46% split after binarizing the target), so accuracy is still a meaningful metric but ROC-AUC / recall are tracked as well.
- `chol` and `trestbps` show weak linear correlation with target individually, but remain clinically relevant and are kept as model features.
- No extreme outliers require row-dropping; StandardScaler in the modeling pipeline handles the differing numeric scales.

## 8. Save Cleaned Dataset

```python
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/heart_clean.csv', index=False)
print('Saved cleaned dataset.')
```
